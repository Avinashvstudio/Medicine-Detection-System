```python
import pandas as pd

# Load the dataset
df4 = df3

# Function to search for medicine
def search_medicine(medicine_name):
    result = df4[df4["name"].str.lower() == medicine_name.lower()]
    if not result.empty:
        return result[["name", "use0", "sideEffect0"]].to_dict(orient="records")
    else:
        return "Medicine not found"

# Test the function
medicine_name = "allegra-m tablet"
print(search_medicine(medicine_name))
```
